```python
import warnings 
warnings.filterwarnings("ignore")
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("titanic.csv")
```

```python
df.shape
```

```python
df.info()
```

```python
df = df.drop(columns = ['PassengerId','Name','Ticket','Cabin','Fare','Embarked','Pclass','SibSp','Parch'])
```

```python
df['Age'].fillna(df['Age'].mean(),inplace = True)
df.isnull().sum()
```

```python
df.info()
```

```python
X = df[["Sex","Age"]]
Y = df["Survived"]
```

```python
Y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size = 0.3)
```

```python
x1_train = pd.get_dummies(x_train,dtype = 'int')
```

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
log_model.fit(x1_train.values,y_train.values)
```

```python
log_model.score(x1_train.values,y_train.values)
```

```python
from sklearn.tree import DecisionTreeClassifier
dec_model = DecisionTreeClassifier()
dec_model.fit(x1_train.values,y_train.values)
```

```python
dec_model_acc = dec_model.score(x1_train.values,y_train.values)
print(f"Accuracy of Decision Tree Model : {dec_model_acc}")
```
